```python
from tkinter import *


def show_entry_fields():

    p1 = str(e1.get())
    p2 = str(e2.get())
    p3 = float(e3.get())
    p4 = float(e4.get())
    p5 = str(e5.get())

    data_new = pd.DataFrame({
     'Make':p1,
        'Model':p2,
        'Year':p3,
        'Engine Size (L)':p4,
        'Fuel Type':p5
         },index=[0])
    result=pipe.predict(data_new)
    Label(master, text=("Price (USD) = ")).grid(row=8)
    Label(master, text=result).grid(row=10)
    print("Price (USD)",result[0])

master=Tk()
master.title("VEHICLE VATICINATOR")
label=Label(master,text="VEHICLE VATICINATOR",bg="black",fg="white").\
                                         grid(row=0,columnspan=2)

Label(master, text="Make").grid(row=1)
Label(master, text="Model").grid(row=2)
Label(master, text="Year").grid(row=3)
Label(master, text="Engine Size (L)").grid(row=4)
Label(master, text="Fuel Type").grid(row=5)


e1=Entry(master)
e2=Entry(master)
e3=Entry(master)
e4=Entry(master)
e5=Entry(master)

e1.grid(row=1, column=1)
e2.grid(row=2, column=1)
e3.grid(row=3, column=1)
e4.grid(row=4, column=1)
e5.grid(row=5, column=1)
       
Button(master, text="Predict",command=show_entry_fields).grid()     

mainloop()
       
```
